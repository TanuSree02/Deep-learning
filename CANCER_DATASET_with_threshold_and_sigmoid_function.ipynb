```python
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split


data = load_breast_cancer()

x = data.data
y = data.target


# convert 0,1 → -1,1

y = np.where(y == 0, -1, 1)


# split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0
)


# normalization

x_min = x_train.min(axis=0)
x_max = x_train.max(axis=0)

x_train = (x_train - x_min) / (x_max - x_min)
x_test = (x_test - x_min) / (x_max - x_min)


np.random.seed(0)

weights = np.random.randn(x_train.shape[1])
bias = 0.0

learning_rate = 0.01
epochs = 5


# threshold function

def threshold(z):
    return 1 if z >= 0 else -1


print("Training phase (Threshold function)")
print("="*60)


for epoch in range(1, epochs+1):

    correct = 0

    for i in range(len(x_train)):

        net_input = np.dot(x_train[i], weights) + bias

        prediction = threshold(net_input)

        error = y_train[i] - prediction

        weights += learning_rate * error * x_train[i]
        bias += learning_rate * error

        if prediction == y_train[i]:
            correct += 1


    epoch_accuracy = (correct / len(x_train)) * 100

    print(f"Epoch {epoch}")
    print(f"Training accuracy : {epoch_accuracy:.2f}%")
    print(f"Bias : {bias:.4f}")
    print("Weights (first 5) :", weights[:5])
    print("-"*60)



# final training accuracy

correct = 0

for i in range(len(x_train)):

    net_input = np.dot(x_train[i], weights) + bias

    prediction = threshold(net_input)

    if prediction == y_train[i]:
        correct += 1


print("Final training accuracy :",
      (correct / len(x_train)) * 100)


# testing

print("\nTesting phase")
print("="*60)

correct = 0

for i in range(len(x_test)):

    net_input = np.dot(x_test[i], weights) + bias

    prediction = threshold(net_input)

    if prediction == y_test[i]:
        correct += 1


print("Final test accuracy :",
      (correct / len(x_test)) * 100)
```
