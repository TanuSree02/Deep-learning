```python
import numpy as np


x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])


w = 0
b = 0

lr = 0.01
epochs = 50


for epoch in range(epochs):

    for i in range(len(x)):

        y_pred = w * x[i] + b

        dw = -2 * x[i] * (y[i] - y_pred)
        db = -2 * (y[i] - y_pred)

        w = w - lr * dw
        b = b - lr * db


print("Weight :", w)
print("Bias :", b)
```
